```python
# Classify Movies Based on Ratings [Good,Better and Best]
# Here we are gonna add another column
def ratingcat(rating):
  if (rating>=7.0):
    return 'Excellent'
  elif(rating>=6.0):
    return 'Good'
  else:
    return 'Average'
df['rating_cat']=df['Rating'].apply(ratingcat)
```
